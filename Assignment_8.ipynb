```python
import numpy as np

arr = np.zeros((10,5,3))
print(arr)
```

```python
#Create another array woth values 255 and shape 10x5x3.
import numpy as np

arr1= np.empty((10,5,3))
arr1[:]=255
print(arr1)
print("\n\ndata type: ",arr1.dtype)
print("\n\ndimension: ",arr1.ndim)
```

```python
# Join the 2 array horizontally

a2h = np.hstack((arr, arr1))
print(a2h)
```

```python
# Create an array of random unsigned integer 8 bits with shape (50x50x3)
# Convert the array from question 4, top part to red and bottom part to green
# Seed value is 11
# Display or show each data

import matplotlib.pyplot as plt

np.random.seed(11)
arr3 = np.random.randint(0,256,(50,50,3),dtype='uint8')
print(arr3)

arr3[:25,:,:] = [255,0,0]
arr3[25:,:,:] = [0,255,0]


print("upper part:\n",arr3)
plt.imshow(arr3)
plt.show()

```
